```python
import pandas as pd
import numpy as np
import random

db = pd.read_csv('Match.csv')
db.info()
```

```python
db.sample(n=20, random_state=0)
```

### __Primero Estandarizamos los atributos numéricos con StandardScaler:__

```python
from sklearn.preprocessing import StandardScaler
from pandas import DataFrame

scaler=StandardScaler()

scaled = scaler.fit_transform(db[['avg_minutes_game','avg_seconds_point','avg_minutes_set','match_minutes']])
print(db[['avg_minutes_game','avg_seconds_point','avg_minutes_set','match_minutes']])
print('\n')
dataset = DataFrame(scaled)
dataset.columns = ['avg_minutes_game','avg_seconds_point','avg_minutes_set','match_minutes']
print('\n')
print(dataset)
```

```python
db1=db.copy()
db1.loc[db1.index.isin(dataset.index), ['avg_minutes_game','avg_seconds_point','avg_minutes_set','match_minutes']] = dataset[['avg_minutes_game','avg_seconds_point','avg_minutes_set','match_minutes']].values
db1.sample(n=20, random_state=0)
```

### __Ahora normalizamos los atributos categóricos en dummy:__

```python
dummy1=pd.get_dummies(db1['tournament'])
dummy1.sample(n=20, random_state=0)
```

```python
dummy2=pd.get_dummies(db['round'])
dummy2.sample(n=20, random_state=0)
```

```python
df=db1.merge(dummy1, left_index=True, right_index=True)
df.sample(n=20, random_state=0)
```

```python
df2=df.merge(dummy2, left_index=True, right_index=True)
df2.sample(n=20, random_state=0)
```
